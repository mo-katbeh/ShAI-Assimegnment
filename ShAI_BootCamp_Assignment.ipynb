**4. Basic Data Visualization:**

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Create a histogram for 'TotalPay'
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Total Pay')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
department_counts = df['JobTitle'].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

**6. Simple Correlation Analysis:** 


```python
columns_for_correlation = ['TotalPay', 'Benefits']

print("Missing values before handling:")
print(df[columns_for_correlation].isnull().sum())

df_cleaned = df.dropna(subset=columns_for_correlation)

print("\nMissing values after handling:")
print(df_cleaned[columns_for_correlation].isnull().sum())

correlation_matrix = df_cleaned[columns_for_correlation].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Scatter plot to visualize the relationship between 'TotalPay' and 'Benefits'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='TotalPay', y='Benefits', color='skyblue')
plt.title('Scatter Plot: Total Pay vs. Benefits')
plt.xlabel('Total Pay')
plt.ylabel('Benefits')
plt.show()
```
